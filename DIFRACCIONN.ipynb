```python
from numpy import cos,sin,pi
from pylab import * 
from  pylab import plot, show

def J(m,x):
    def f(teta):
        return cos(m*teta - x*sin(teta))
    
    N = 1000
    a = 0.
    b = pi
    h = (b-a)/N
    
    s = f(a) + f(b) + 4*f(b-h)
    for k in range(1,N//2):
        s += 4*f(a + (2*k-1)*h) + 2*f(a+2*k*h)
    
    I = h/3*s/pi

    return I

x = linspace(0,20)
grid()
plt.title("FUNCIONES DE BESSEL")   # Establece el título del gráfico
#plt.xlabel("Longtud de Onda")   # Establece el título del eje x
#plt.ylabel("Intensidad")   # Establece el título del eje y
plot(x,J(0,x),label='J0', color='k')
plot(x,J(1,x),label='J1', color='g')
plot(x,J(2,x),label='J2', color='y')
legend()
```
